```python
# Solicitar los números al usuario
num1 = int(input("Ingresa el primer número: "))
num2 = int(input("Ingresa el segundo número: "))

# Encontrar el menor de los dos números para limitar el ciclo
if num1 < num2:
    menor = num1
else:
    menor = num2

# Inicializar una variable para almacenar el MCD
mcd = 1

# Ciclo for para encontrar el mayor divisor común
for i in range(1, menor + 1):
    if num1 % i == 0 and num2 % i == 0:
        mcd = i

# Imprimir el resultado
print(f"El máximo común divisor de {num1} y {num2} es {mcd}")
```

```python
import pandas as pd

def calcular_estadisticas(descargas: pd.DataFrame) -> pd.DataFrame:
    # Filtrar los modelos con al menos un pago mayor a 0
    descargas = descargas[descargas['PAGO'] > 0]

    # Definir la función para calcular las estadísticas por grupo
    def calcular_grupo(grupo):
        return pd.Series({
            'CANTIDAD': grupo['PAGO'].count(),
            'PROMEDIO': grupo['PAGO'].mean(),
            'MAXIMO': grupo['PAGO'].max(),
            'MINIMO': grupo['PAGO'].min(),
            'ESTRELLAS': grupo['ESTRELLAS'].mean(),
            'DESV_ESTRELLAS': grupo['ESTRELLAS'].std(ddof=0) if len(grupo['ESTRELLAS']) > 1 else 0.0,
            'COMENTARIOS': grupo['COMENTARIO'].sum()
        })

    # Aplicar la función a cada grupo
    estadisticas = descargas.groupby('MODELO').apply(calcular_grupo)

    # Redondear las columnas numéricas a dos decimales
    estadisticas = estadisticas.round(2)

    # Ordenar los resultados alfabéticamente por el nombre del modelo
    estadisticas = estadisticas.sort_index()

    return estadisticas
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests

# Función para extraer datos de acciones utilizando yfinance
def get_stock_data(ticker):
    stock = yf.Ticker(ticker)
    data = stock.history(period="1y")
    return data

# Función para graficar datos de acciones
def plot_stock_data(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.show()

# Función para realizar web scraping de ingresos
def get_revenue_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    tables = soup.find_all('table')  # Encuentra todas las tablas en la página
    if tables:
        df = pd.read_html(str(tables[0]))[0]
        return df
    else:
        raise ValueError("No tables found at the provided URL.")

# Extracción de datos de acciones de Tesla
tesla_stock_data = get_stock_data(ticker)
print(tesla_stock_data)
plot_stock_data(tesla_stock_data, 'Tesla Stock Price (Last Year)')

# Extracción de datos de ingresos de Tesla (Ejemplo: URL real de Yahoo Finance)
tesla_revenue_url = 'https://finance.yahoo.com/quote/TSLA/financials?p=TSLA'
try:
    tesla_revenue_data = get_revenue_data(tesla_revenue_url)
    print(tesla_revenue_data)
except ValueError as e:
    print(e)

# Extracción de datos de acciones de GameStop
gamestop_stock_data = get_stock_data('GME')
print(gamestop_stock_data)
plot_stock_data(gamestop_stock_data, 'GameStop Stock Price (Last Year)')

# Extracción de datos de ingresos de GameStop (Ejemplo: URL real de Yahoo Finance)
gamestop_revenue_url = 'https://finance.yahoo.com/quote/GME/financials?p=GME'
try:
    gamestop_revenue_data = get_revenue_data(gamestop_revenue_url)
    print(gamestop_revenue_data)
except ValueError as e:
    print(e)

# Creación de un tablero para Tesla
def create_dashboard(stock_data, revenue_data, title):
    fig, axs = plt.subplots(2, 1, figsize=(10, 10))
    fig.suptitle(title)

    axs[0].plot(stock_data['Close'])
    axs[0].set_title('Stock Price')
    axs[0].set_xlabel('Date')
    axs[0].set_ylabel('Close Price')

    if revenue_data is not None:
        axs[1].table(cellText=revenue_data.values, colLabels=revenue_data.columns, loc='center')
        axs[1].set_title('Revenue Data')
        axs[1].axis('off')
    else:
        axs[1].text(0.5, 0.5, 'No revenue data available', horizontalalignment='center', verticalalignment='center', fontsize=12)
        axs[1].axis('off')

    plt.show()

create_dashboard(tesla_stock_data, tesla_revenue_data if 'tesla_revenue_data' in locals() else None, 'Tesla Dashboard')

# Creación de un tablero para GameStop
create_dashboard(gamestop_stock_data, gamestop_revenue_data if 'gamestop_revenue_data' in locals() else None, 'GameStop Dashboard')

# Guardar el cuaderno de trabajo
tesla_stock_data.to_csv('tesla_stock_data.csv')
if 'tesla_revenue_data' in locals():
    tesla_revenue_data.to_csv('tesla_revenue_data.csv')
gamestop_stock_data.to_csv('gamestop_stock_data.csv')
if 'gamestop_revenue_data' in locals():
    gamestop_revenue_data.to_csv('gamestop_revenue_data.csv')
```

```python
#SMA 5 dias
df['SMA_short']=df['Adj Close'].rolling(window=5).mean()
#SMA 20 dias
df['SMA_long']=df['Adj Close'].rolling(window=50).mean()
df
```

```python
plt.figure(figsize=(16,8))
plt.title('AAPL, SMA-5 & SMA 20 days')
plt.plot(df['Adj Close'])
plt.plot(df['SMA_short'],color='orange',alpha=0.4)
plt.plot(df['SMA_long'],color='red',alpha=0.4)
plt.xlabel('fecha')
plt.ylabel('USD $')
plt.legend(['Adj close','SMA_short','SMA_long'])
plt.show()
```

```python
#media movil exponencial 5 dias
df['EMA_short']=df['Adj Close'].ewm(span=5,adjust=False).mean()
#media movil exponencial 20 dias
df['EMA_long']=df['Adj Close'].ewm(span=20,adjust=False).mean()
df
```

```python
plt.figure(figsize=(16,8))
plt.title('AAPL, SMA-5 & SMA 20 days')
plt.plot(df['Adj Close'])
plt.plot(df['EMA_short'],color='orange',alpha=0.4)
plt.plot(df['EMA_long'],color='red',alpha=0.4)
plt.xlabel('fecha')
plt.ylabel('USD $')
plt.legend(['Adj close','EMA_short','EMA_long'])
plt.show()
```

```python
#bandas de bollinger
bb=ta.volatility.BollingerBands(close=df['Close'],window=50,window_dev=2)

df['BOL_M']=bb.bollinger_mavg()

df['BOL_U']=bb.bollinger_hband()

df['BOL_D']=bb.bollinger_lband()

df
```

```python
plt.figure(figsize=(16,8))
plt.title('APPL,bolliger bands 20 days')
plt.plot(df['Adj Close'])
plt.plot(df['BOL_M'],color='orange',alpha=0.4)
plt.plot(df['BOL_U'],color='red',alpha=0.4)
plt.plot(df['BOL_D'],color='blue',alpha=0.4)
plt.xlabel('fecha')
plt.ylabel('USD $')
plt.legend(['Adj close','BOL_M','BOL_U','BOL_D'])
plt.show()
```

```python
#media movil de convergencia
df['MACD']=ta.trend.MACD(close=df['Close'],window_fast=12,window_slow=26).macd()
df
```

```python
plt.figure(figsize=(16,8))
plt.title('AAPL,MACD(12,26)')
plt.plot(df['MACD'],color='blue',alpha=0.4)
plt.xlabel('fecha')
plt.ylabel('MACD')
plt.axhline(y=0,color='green')
plt.show()
```

```python
#retrocesos de Fibonacci
max_p=df['Close'].max()
min_p=df['Close'].min()
dif_p=max_p-min_p
print(max_p)
print(min_p)
print(dif_p)
r_fibo=np.array([0.236,0.382,0.5,0.618])
levels=max_p-dif_p*r_fibo
levels
```

```python
plt.figure(figsize=(16,8))
plt.title('AAPL,Fibonacci Levels')
plt.plot(df['Close'],color='blue',alpha=0.4)
plt.xlabel('fecha')
plt.ylabel('USD $')
plt.axhline(y=max_p,color='green',linestyle='--',alpha=0.6)
plt.axhline(y=levels[0],color='green',linestyle='--',alpha=0.6)
plt.axhline(y=levels[1],color='blue',linestyle='--',alpha=0.6)
plt.axhline(y=levels[2],color='purple',linestyle='--',alpha=0.6)
plt.axhline(y=levels[3],color='red',linestyle='--',alpha=0.6)
plt.axhline(y=min_p,color='green',linestyle='--',alpha=0.6)

plt.show()
```

```python
def generar_senal(df):
    senales = []
    en_posicion = False

    for i in range(len(df)):
        if df['Close'][i] > df['SMA_long'][i] and df['Close'][i] > df['BOL_U'][i]:
            if not en_posicion:
                senales.append("Compra")
                en_posicion = True
            else:
                senales.append("Espera")
        elif df['Close'][i] < df['SMA_long'][i] and df['Close'][i] < df['BOL_D'][i]:
            if en_posicion:
                senales.append("Venta")
                en_posicion = False
            else:
                senales.append("Espera")
        else:
            senales.append("Espera")

    return senales

señales = generar_senal(df)
df['Señales'] = señales
```
