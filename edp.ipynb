# **Importamos las bibliotecas necesarias**

```python

import numpy as np
import matplotlib.pyplot as plt
```

# **Parámetros del Problema**

```python
L = 2        # Longitud del dominio en x
T = 1        # Tiempo máximo
Nx = 20      # Número de puntos en x
Nt = 100     # Número de puntos en t
dx = L / (Nx - 1)  # Tamaño del paso en x
dt = T / Nt        # Tamaño del paso en t
alpha = dt / dx**2 # Coeficiente de estabilidad
```

# **Creamos la malla de x y t**

```python
x = np.linspace(0, L, Nx)
t = np.linspace(0, T, Nt)
```

# **Condición inicial**

```python
u = np.zeros((Nx, Nt))
for i in range(1, Nx-1):
    u[i, 0] = np.sin(np.pi * x[i] / 2)
```

# **Método de Diferencias Regresivas para resolver la EDP**

```python
for j in range(0, Nt - 1):
    for i in range(1, Nx - 1):
        u[i, j+1] = u[i, j] + alpha * (u[i+1, j] - 2 * u[i, j] + u[i-1, j])
```

# **Visualización de la solución**

```python
plt.figure(figsize=(8, 5))
for j in range(0, Nt, Nt // 5):  # Ploteamos cada cierto número de pasos de tiempo
    plt.plot(x, u[:, j], label=f't = {t[j]:.2f}')
plt.xlabel("x")
plt.ylabel("u(x, t)")
plt.title("Solución de la EDP usando Diferencias Regresivas")
plt.legend()
plt.grid(True)
plt.show()
```
