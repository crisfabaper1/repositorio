```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Configuracin de la figura y el eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Parmetros de la onda
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Inicializacin de la funcin de la onda
def wave_function(X, Y, t):
    Z = np.sin(np.sqrt(X**2 + Y**2) - t)
    return Z

# Crear la superficie inicial de la onda
Z = wave_function(X, Y, 0)
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Limitar el eje z para visualizar bien la onda
ax.set_zlim(-1, 1)

# Funcin de actualizacin para la animacin
def update(t):
    ax.clear()
    Z = wave_function(X, Y, t)
    surf = ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_zlim(-1, 1)
    return surf,

# Crear la animacin
ani = FuncAnimation(fig, update, frames=np.linspace(0, 20, 100), interval=50)

# Mostrar la animacin
plt.show()
```
